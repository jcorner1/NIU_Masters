```python
ds = xr.open_mfdataset('/home/scratch/20CR_v3/hgt.*.nc', parallel=True)
daily_z500_mean = ds.hgt.sel(level=500., time=slice('1979-01-01', '2015-12-31')).groupby('time.dayofyear').mean(dim='time')
z500_mean = daily_z500_mean.compute()
z500_mean.to_netcdf('z500_mean_dayofyear_1979_2015.nc')
```

```python
daily_z500_stdev = ds.hgt.sel(level=500., time=slice('1979-01-01', '2015-12-31')).groupby('time.dayofyear').std(dim='time')
z500_std = daily_z500_stdev.compute()
z500_std.to_netcdf('z500_stdev_dayofyear_1979_2015.nc')
```
